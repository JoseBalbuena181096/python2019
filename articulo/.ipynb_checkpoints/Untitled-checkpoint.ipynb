```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import matplotlib.image as mpimg
```

```python
images_src=[]
for i in range(0,4):
    img = mpimg.imread("../articulo/zero/sample"+str(i)+".jpg")
    images_src.append(img)
```

```python
images_src1=[]
for i in range(0,4):
    img = mpimg.imread("../articulo/one/sample"+str(i)+".jpg")
    images_src1.append(img)
```

```python
images_src2=[]
for i in range(0,4):
    img = mpimg.imread("../articulo/two/sample"+str(i)+".jpg")
    images_src2.append(img)
```

```python
images_src3=[]
for i in range(0,4):
    img = mpimg.imread("../articulo/three/sample"+str(i)+".jpg")
    images_src3.append(img)
```

```python
def plot_image(imgs,name):
    fig, axs = plt.subplots(2, 2)
    k = 0
    for i in range(0,2):
        for j in range(0,2):
            if k == 0:
                t = "Straight"
            elif k == 1:
                t = "Curve A"
            elif k == 2:
                t = "Curve B"
            elif k == 3:
                t = "Crossing"
            axs[i, j].set_title(t)
            axs[i,j].imshow(imgs[k])
            k += 1
    for ax in axs.flat:
        ax.set(xlabel="column", ylabel="row")
    for ax in axs.flat:
        ax.label_outer()
    plt.savefig(name+'.png')
```

```python
plot_image(images_src,"zero")
```

```python
plot_image(images_src1,"one")
```

```python
plot_image(images_src2,"two")
```

```python
plot_image(images_src3,"three")
```

```python
data = pd.read_csv("../articulo/data.csv")
mean = data.mean(axis = 0) 
for i in range(0,6):
    data["col_data"+str(i)].fillna(mean[i], inplace=True)
data.head()
```

```python
def distribution_variable(df,col,c,name):
    fig, axs = plt.subplots(2, 3)
    k = 0
    t = ""
    for i in range(0,2):
        for j in range(0,3):
            if k == 0:
                t = "Straight left"
            elif k == 1:
                t = "Curve A left"         
            elif k == 2:
                t = "Curve B left"    
            elif k == 3:
                t = "Straight right"  
            elif k == 4:
                t = "Curve B right"
            elif k == 5:
                t = "Curve B right"
            axs[i, j].set_title(t)
            sns.distplot(df[col+str(k)], hist=False, color=c, kde_kws={"shade": True}, ax=axs[i,j])
            k +=1
    for ax in axs.flat:
        ax.set(xlabel="column",ylabel="Density")
    for ax in axs.flat:
        ax.label_outer()
    plt.savefig(name+'.png')
```

```python
distribution_variable(data,"col_data","g","left")
```

```python
def distribution_variable(df,col,c,name):
    fig, axs = plt.subplots(1, 3)
    k = 0
    t = ""
    for j in range(0,3):
        if k == 0:
            t = "Straight"
        elif k == 1:
            t = "Curve A"
        elif k == 2:
            t = "Curve B"
        elif k == 3:
            t = "Crossing"
        axs[j].set_title(t+" "+name)
        sns.distplot(df[col+str(k)], hist=False, color=c, kde_kws={"shade": True}, ax=axs[j])
        k +=1
    for ax in axs.flat:
        ax.set(xlabel="column",ylabel="Density")
    for ax in axs.flat:
        ax.label_outer()
    plt.savefig(name+'.png')
```

```python
distribution_variable(data,"col_data","g","left")
```
