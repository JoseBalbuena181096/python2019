# Error absoluto medio 
El Error Absoluto Medio(Mean Absolute Error o MAE) se define como:
$	\frac{1}{n} \sum_{i=1}^{n}|y_{i}\widehat{y_{i}}|$

Es decir, la media de las diferencias entre la variable objetivo y las predicciones sin signo. 
MAE es una metrica robusta, encuanto a que no varia mucho si hay valores extremos de datos. El error se puede interpretar como unidades de la variable objetivo (si la variable objetivo es en dolares MAE esta tambien en dolares).

```python
mean_absolute_error = data["error_absoluto"].sum()/data["error_absoluto"].count()
mean_absolute_error
```

# Error cuadratico medio 
El Error Cuasratico Medio(Mean Squared Error o RMSE) se deferencia del MSE su redultado se mide en las mismas unidades que el objetivo:
$	\sqrt{\frac{1}{n} \sum_{i=1}^{n}(y_{i}-\widehat{y_{i}})^2}$
