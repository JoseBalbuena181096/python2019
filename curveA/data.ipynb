```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_left=[]
for i in range(0,6):
    source = "../curveA/csv/dataL"+str(i)+".csv"
    aux = pd.read_csv(source)
    data_left.append(aux)
```

```python
data_left
```

```python
def plot_data(df,col_1,col_2):
    df.plot.scatter(x=col_2,y=col_1)
    plt.show()
```

```python
for i in range(0,6):
    plot_data(data_left[i],"row_data","col_data")
```

```python
for i in range(0,6):
    plot_data(data_left[i],"row_model","col_model")
```

```python
def distribution_variable(df,col):
    df[col].plot.kde()
    plt.xlabel("Variable distribution {}".format(col))
    plt.show()
```

```python
for i in range(0,6):
    distribution_variable(data_left[i],"col_data")
```

```python
for i in range(0,6):
    distribution_variable(data_left[i],"col_model")
```

```python
for i in range(0,6):    
    data_left[i]["absolute_error"]= abs(data_left[i]["col_model"]-data_left[i]["col_data"])
```

```python
data_left
```

```python
MAE_LEFT = []
for i in range(0,6):    
    aux = data_left[i]["absolute_error"].sum()/data_left[i]["absolute_error"].count()
    MAE_LEFT.append(aux)
MAE_LEFT
```

```python
for i in range(0,6):    
    data_left[i]["quadratic_error"]= pow((data_left[i]["col_model"]-data_left[i]["col_data"]),2)
```

```python
data_left
```

```python
MSE_LEFT = []
for i in range(0,6):    
    aux =  np.sqrt(data_left[i]["quadratic_error"].sum()/data_left[i]["quadratic_error"].count())
    MSE_LEFT.append(aux)
MSE_LEFT 
```

```python
left_errors = pd.DataFrame()
left_errors["MSE_LEFT"] = MSE_LEFT 
left_errors["MAE_LEFT"] = MAE_LEFT
left_errors
writer = pd.ExcelWriter('../curveA/errorL.xlsx')
left_errors.to_excel(writer)
writer.save()
```

```python
distribution_variable(left_errors,"MAE_LEFT")
```

```python
distribution_variable(left_errors,"MSE_LEFT")
```
