# Convolution:
## Intuition
Convolutions use a kernel matrix to scan a given image and apply a filter to obtain  a certain effect.

A image kernel is a matrix used to apply effects such as blurring and sharpening.

Kernels are used in marchine lerning for future extraction to select most import pixels of an image.
 
Convolution preserves the spatial relationship  between pixels.

### Gaussian blur kerner approximation (5,5)
![image info](./notebook/Gaussian_Kernel.png)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("test_image.jpg")
lane_image=np.copy(image)
gray=cv2.cvtColor(lane_image,cv2.COLOR_RGB2GRAY)
#apply a 5x5 gaussian blur to gray image
blur=cv2.GaussianBlur(gray,(5,5),0)
plt.imshow(gray)
```
