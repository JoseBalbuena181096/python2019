# The hough Transform
## The basic .
The hough tranasfornd is a incredible tool that lets you identify lines. Not just lines, but other shapes as well in this article, I´ll talk about the mechanics behind the Hough transform.
## Why the hough trasnsform?
Lets say you the snapshot of pole. You figure out edge pixels (using the Canny edge detector, the Sobel edge detector o any other thing). Now you want a geometrical representation of the pole´s edge. You want to know its slope ,its intercept, etc. But right now the "edge" is just a sequence of pixels.
You can loop through all pixels, and some how figure out the slope and intercept. But that is one difdicult task. Image naver perfect.

So you want some mechanism that give more weightage to pixels that are already in a line. This is exactly what the Hough transform does.
It lest each point on the image "vote".And because of the mathematical propietes of the transform, this "voting" allows us to figure out prominent lines in the image.
## From lines to points.
A lines is a collection of points. And managing a collection of points is tougher than managing a single point. Obviously. So the first thing we learn is how represent a line as a single point, without losing any information about it.
This is done through the m-c space.
![image info](./notebook/hough1.jpg)
As shown in the above picture, every line has two quantities associated with it, the slope and the intercept. With these two numbers, you can describe a line completely.
So the parameter space, or the mc space was devised.Every line in the xy space is equivalent to a single point in the mc space.
## From points to lines
Now onto the next step. Consider a point(say($x_{a},y_{a} $)) in the xy space. What would its representation in the mc space be?
For starters, you could guess that infinite lines pass through a point. So, for every line passing throungh ($x_{a},y_{a}$), there would be a point in the mc space.
1. And you are correct there through $(x_{a},y_{a}):=mx_{a}+c$.
2.Rearranging:$c=-x_{a}m+y_{a}$
3. The above is the equation of a line in the mc space.
So, a point in the xy space is equivalent to a line in the mc space.
## How does help us?
The hough transform is all about doing what we just learned: converting points in the xy spcace to lines in the mc space.
You taken an image an edge detected image, andd for every point that is not black, you draw lines in the mc place. Obviously, some lines will be intersect. These intersections mark are the parameters of the line.
The following picture will clarify the idea:
![image info](./notebook/hough2.jpg)
The ponints 1,2,3 and 4 are represend as various lines in the mc space. And the intersection of these lines is equivalent to the original line.
# The normal form
The flaw
The Hough transform describe in the previous has a flaw. The value of m (slope) tehds to infinite for vertical lines. So you need infinite memory to be able to store the mc space. Not good.
## The solution 
The problem is resolved by using a different parametrization, instead of the slope-intercept form of lines, we use the normal form.
![image info](./notebook/hough3.jpg)
In this representation, a line is formed using two parameters -an angle $\theta$ and a distance $\rho$. $\rho$ is the lengh of the normal from the origin(0,0) onto the line, and $\theta$ is the angle this nolmal makes with the x axis. This solves the flaw perfectly
The angle $\theta$ can be only from -90° to 90° and the lenght $\rho$ can range from 0 to the length of diagonal of the image. These value are finete, and thus you can "store" them on a computer without much trouble.
In this representation, the equation of the line is  
$\rho=x_{1}cos(\theta)+y_{1}sin(\theta)$
where ($x_{1},y_{1}$) is a point throght which the line passes.
With this new equation, we have a few changes in the end-point to line "transition" from the xy space to the $\rho \theta$  space. A line in the xy space is still equivalent to a point in the $\rho\theta$ space. But a point xy space is now equivalent to a sinusoidal curve have been generated while trying to figure out lines an image. 
## Implemetation 
With those ideas through we are ready to implement 
the hough transform. The idea is to let each pixel "vote". So array of accumulator cell is created.
In our case, the accumilator cells form a 2D array. The horizontal axis is for the different  $\theta$ values and the vertical axis if for $\rho$ values.
![image info](./notebook/hough4.jpg)
Next, you loop through every pixel of the edge detected image(Hough only work only on edge detected image...not on normal image).
If a pixel is zero, you ignore it, it´s not an edge, so it can´t be a line. So move on to the next pixel.
If a pixel is nonzero, you generate its sinusoidal curve(in the $\rho\theta$ space). That is, you take $\theta=-90$ and calculate the corresponding $\rho$ value. Then you "vote " in the accumulator cell ($\theta,\rho$). That is you increase the value of this cell by 1. Then you taje the next $\theta$ value and calculate the next $\rho$ value. And so on till $\theta$ =+90. And for every such calculation making sure they "vote".
So for every nonzero pixel, you will get a sinusoidal curve in the $\rho\theta$ space (the accumulator cells). And you will end up with an image similar to the one at the top.
### Detecting lines.
The image at the top has several "bright spots". A lot of points "voted" for this spot. And these are the parameters that describe the lines in the original image. Simple as that.
### Accuracy
The accuracy of the hough tramsform depends on the number of accumulator cells you have. Say you have only -90°,-45°,0,45°,90° as the cells for $\theta$ values. The "voting" process would be terrible inaccurate. Similary for the $\rho$ axis. The more cells you have along a particular axis, the more accurate the transform would be.
Also, the tecnique depends on several "votes" being cast into small region. Only then would you get those bright spots. Otherwise, differentiating between a line and background noise is one really tough job.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def canny(image):
    gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY);
    blur=cv2.GaussianBlur(gray,(5,5),0)
    canny=cv2.Canny(blur,50,150)
    return canny
def region_of_interest(image):
    height=image.shape[0]
    polygons=np.array([
        (200,height),(1100,height),(550,250)
    ])
    mask=np.zeros_like(image)
    #fillPoly(image_destinity,area_apply,intensity)
    cv2.fillPoly(mask,[polygons],255)
    #apply the operation AND between the mask and canny image
    masked_image=cv2.bitwise_and(image,mask)
    return masked_image
def display_lines(image,lines):
    line_image=np.zeros_like(image)
    if lines is not None:
        for line in lines:
            x1,y1,x2,y2=line.reshape(4)
            cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),10)
    return line_image      
            
```

Probabilistic Hough Line Transform.
HoughLinesP(dst,lines,1,CV_PI/180,50,50,10);
with the arguments:
- dst: Output of the edge detector. It should be a grayscale image(although in fact it is a binary one).
- lines: A vector that will store the parameters($X_{start},Y_{start},X_{end},Y_{end}$) of the detected lines.
- rho: the resolution of the parameter $\rho$ in pixels. This use a pixel.
- theta: The resolution of the parameter $\theta$ in radians. We use 1 degree(CV_PI/180).
- thereshold: The minimun number of intersection to "detect" a line.
- minLinLength: The minumum number of pints than can form a line.Lines with less than this number of points is disregarded.
- maxLineGap: The maximum gap between two points to be considered in the same line.

Each line is a 2D array containing our line coordinates in the form[[x1,y1,x2,y2]]. These coordinates specify the line´s parameters, as well as the localitation of the lines with respect to the image space, ensuring that they are placed in the correct position.

```python
image=cv2.imread("test_image.jpg")
lane_image=np.copy(image)
canny_image=canny(lane_image)
cropped_image=region_of_interest(canny_image)
#Apply Probabilist Hough Transform
#cv2.HoughLineP(image_cany,rho_acuracie,theta_acuracie,array_for_output,minLineLength,maxLineGap)
lines=cv2.HoughLinesP(cropped_image,2,np.pi/180,100,np.array([]),minLineLength=40,maxLineGap=5)
line_image=display_lines(lane_image,lines)
#add line to color image
combo_image=cv2.addWeighted(lane_image,0.8,line_image,1,1)
plt.imshow(line_image)
plt.show()
```
